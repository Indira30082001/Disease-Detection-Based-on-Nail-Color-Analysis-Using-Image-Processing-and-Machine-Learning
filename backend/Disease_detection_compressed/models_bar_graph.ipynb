```python
import matplotlib.pyplot as plt

# Models and their accuracies
models = ['RF', 'Cnn', 'Knn', 'XGboost']
accuracies = [0.5, 0.6, 0.2, 0.5]

# Create the bar graph
plt.bar(models, accuracies, color='teal')

# Add labels and title
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Model vs Accuracy')

# Show the graph
plt.show()
```
